```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation
plt.style.use('seaborn-pastel')
```

```python
rd = pd.read_csv("raw_datasets/Covid.csv")
pd.to_datetime(rd["dateRep"])
rd = rd[::-1]
rd.set_index("dateRep")
```

```python

rdRus = rd[rd["countriesAndTerritories"] == "Russia"]
rdFr = rd[rd["countriesAndTerritories"] == "France"]
rdGer = rd[rd["countriesAndTerritories"] == "Germany"]
rdItaly = rd[rd["countriesAndTerritories"] == "Italy"]
rdUS = rd[rd["countriesAndTerritories"] == "United_States_of_America"]
rdChina = rd[rd["countriesAndTerritories"] == "China"]
rdIndia = rd[rd["countriesAndTerritories"] == "India"]
count = {
    "Russia" : [x for x in rdRus["cases"]],
    "France" : [x for x in rdFr["cases"]],
    "Germany" : [x for x in rdGer["cases"]],
    "Italy" : [x for x in rdItaly["cases"]],
    "USA" : [x for x in rdUS["cases"]],
    "China" : [x for x in rdChina["cases"]],
    "India" :[x for x in rdIndia["cases"]]
}

cbc = pd.DataFrame(data = count)
cbc = cbc.set_index(rdRus["dateRep"])
```

```python

color = ['red', 'green', 'blue', 'orange', "black", "cyan", "purple"]
fig = plt.figure(figsize = (12, 7))
plt.xticks(rotation=45, ha="right", rotation_mode="anchor") #rotate the x-axis values
plt.subplots_adjust(bottom = 0.2, top = 0.9) #ensuring the dates (on the x-axis) fit in the screen
plt.ylabel('No of Cases')
plt.xlabel('Date')

def buildmelinechart(i=int):
    plt.legend(cbc.columns)
    p = plt.plot(cbc[:i].index, cbc[:i].values) #note it only returns the dataset, up to the point i
    for i in range(0,7):
        p[i].set_color(color[i]) #set the colour of each curve
import matplotlib.animation as ani
animator = ani.FuncAnimation(fig, buildmelinechart, interval = 50)
plt.show()
```
